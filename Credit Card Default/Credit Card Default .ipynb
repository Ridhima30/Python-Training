```python
df=ccd.copy()
```

```python
df
```

```python
df["EDUCATION"].value_counts()
```

```python
#replacing values 0,5,6 with others(val=4)
df["EDUCATION"].replace(0, 4, inplace=True)
df["EDUCATION"].replace(5, 4, inplace=True)
df["EDUCATION"].replace(6, 4, inplace=True)
```

```python
df["EDUCATION"].value_counts()
```

```python
df["SEX"].value_counts()
```

```python
df["MARRIAGE"].value_counts()
```

```python
df["MARRIAGE"].replace(0, 3, inplace=True)
```

```python
df["MARRIAGE"].value_counts()
```

```python
df["AGE"].value_counts()
```

```python
df["PAY_0"].value_counts()
```

```python
# calculate outstanding amount for each month
df["OUT_1"]=df["BILL_AMT1"]-df["PAY_AMT1"]
df["OUT_2"]=df["BILL_AMT2"]-df["PAY_AMT2"]
df["OUT_3"]=df["BILL_AMT3"]-df["PAY_AMT3"]
df["OUT_4"]=df["BILL_AMT4"]-df["PAY_AMT4"]
df["OUT_5"]=df["BILL_AMT5"]-df["PAY_AMT5"]
df["OUT_6"]=df["BILL_AMT6"]-df["PAY_AMT6"]
```

```python
df
```

```python
df["Total OUT"]=df["OUT_1"]+df["OUT_2"]+df["OUT_3"]+df["OUT_4"]+df["OUT_5"]+df["OUT_6"]
```
